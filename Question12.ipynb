**2) Running an IPFS Node Behind a Firewall with a Public IP**

To make my local IPFS node reachable behind a firewall, we need to tunnel the traffic through a remote server with a public IP.

**Set Up a Remote Server:**

Use a cloud service like AWS, DigitalOcean, Linode, or Hetzner to create a VPS (Virtual Private Server) with a public IP.

**Now I Install & Start IPFS on the Remote Server**

SSH into the server:

    *    ssh user@192.168.1.104:4001

Install IPFS:

    *    wget https://dist.ipfs.tech/go-ipfs/v0.22.0/go-ipfs_v0.22.0_linux-amd64.tar.gz

    *    tar -xvzf go-ipfs_v0.22.0_linux-amd64.tar.gz

    *    cd go-ipfs

    *    sudo bash install.sh

Initialize IPFS:

    *    ipfs init

Start IPFS daemon:

    *    ipfs daemon


**Set Up an SSH Tunnel**

On my local machine, tunnel traffic through the remote server:

    *    ssh -L 4001:localhost:4001 -N user@192.168.1.104:4001

This will forward local IPFS traffic through the remote server.


**Announce Public Address**

On the remote server, I edit ~/.ipfs/config and update Addresses to:

    "Addresses": {
      "API": "/ip4/192.168.1.104/tcp/5001",
      "Gateway": "/ip4/192.168.1.104/tcp/8080",
      "Swarm": [
        "/ip4/192.168.1.104/tcp/4001",
        "/ip6/::/tcp/4001"
      ]
    }

i add this json to update the Address.

Now Restart IPFS:

    *    ipfs shutdown

    *    ipfs daemon --enable-pubsub-experiment


**Access My Node from a Public Gateway**

https://192.168.1.104:8080/ipfs/QmXkYhLk7Wz6W5zEgPpZiB5SHf9C6oZhTqZxZZ7B5Ykhds

My locally pinned file is now accessible globally via the remote IPFS node.
