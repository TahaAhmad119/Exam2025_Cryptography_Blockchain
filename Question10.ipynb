**1. Install Required Dependencies**

1st i install Node.js and Hardhat

Then create Hardhat project using these command:

    *    mkdir erc20-tokens && cd erc20-tokens
    *    npx hardhat
    *    npm install --save-dev @nomicfoundation/hardhat-toolbox dotenv

**2. Configure Hardhat for Sepolia**

now i create a enc file add these keys

    *    ALCHEMY_API_KEY=your-alchemy-api-key
    *    PRIVATE_KEY=your-metamask-private-key

then I Edit hardhat.config.js this file place this code in this:

    require("@nomicfoundation/hardhat-toolbox");
    require("dotenv").config();

    module.exports = {
    solidity: "0.8.19",
    networks: {
    sepolia: {
      url: `https://eth-sepolia.g.alchemy.com/v2/${process.env.ALCHEMY_API_KEY}`,
      accounts: [`0x${process.env.PRIVATE_KEY}`],
      },
     },
    };


**3. Write the ERC-20 Token Contract**


Inside contracts/Token1.sol, I create my first ERC-20 token:

    // SPDX-License-Identifier: MIT
    pragma solidity ^0.8.19;

    import "@openzeppelin/contracts/token/ERC20/ERC20.sol";

    contract Token1 is ERC20 {
    constructor() ERC20("Token One", "TK1") {
        _mint(msg.sender, 1000000 * 10 ** decimals());
     }
    }

Similarly, create contracts/Token2.sol for your second token:

    // SPDX-License-Identifier: MIT
    pragma solidity ^0.8.19;

    import "@openzeppelin/contracts/token/ERC20/ERC20.sol";

    contract Token2 is ERC20 {
    constructor() ERC20("Token Two", "TK2") {
        _mint(msg.sender, 1000000 * 10 ** decimals());
     }
    }

**4. Write Deployment Script**

Now i add some code Inside scripts/deploy.js this file:

    async function main() {
      const [deployer] = await ethers.getSigners();
      console.log("Deploying contracts with account:", deployer.address);

      // Deploy Token1
      const Token1 = await ethers.getContractFactory("Token1");
      const token1 = await Token1.deploy();
      await token1.deployed();
      console.log("Token1 deployed to:", token1.address);

      // Deploy Token2
      const Token2 = await ethers.getContractFactory("Token2");
      const token2 = await Token2.deploy();
      await token2.deployed();
      console.log("Token2 deployed to:", token2.address);
    }

    main()
      .then(() => process.exit(0))
      .catch((error) => {
        console.error(error);
        process.exit(1);
      });

**5. Compile & Deploy**

Now I Compile the Contracts using this command:

    *    npx hardhat compile

After this Deploy on Sepolia using this command:

    *    npx hardhat run scripts/deploy.js --network sepolia
