```python
import hashlib
from Crypto.Hash import RIPEMD

# Part a: Find which name hashes to the given SHA-1 hash
def part_a():
    names = ["Mariusz", "Sarah", "Fabian", "Daniel"]
    target_hash = "3b472fb496e7185001072f414b34f91b252a982e"

    for name in names:
        sha1_hash = hashlib.sha1(name.encode()).hexdigest()
        if sha1_hash == target_hash:
            print(f"a) The name that hashes to {target_hash} (SHA-1) is: {name}")
            break

# Part b: Find which name hashes to the given MD5 hash
def part_b():
    names = ["Mariusz", "Sarah", "Fabian", "Daniel"]
    target_hash = "262031397020fd8df478ec13b4b096c5"

    for name in names:
        md5_hash = hashlib.md5(name.encode()).hexdigest()
        if md5_hash == target_hash:
            print(f"b) The name that hashes to {target_hash} (MD5) is: {name}")
            break

# Part c: Compute the SHA-256 hash of "Mariusz"
def part_c():
    name = "Mariusz"
    sha256_hash = hashlib.sha256(name.encode()).hexdigest()
    print(f"c) SHA-256 hash of {name}: {sha256_hash}")

# Part d: Compute the RIPEMD-160 hash of "Fabian"
def part_d():
    name = "Fabian"
    ripemd160_hash = RIPEMD.new(name.encode()).hexdigest()
    print(f"d) RIPEMD-160 hash of {name}: {ripemd160_hash}")

# Part e: Determine the length in bits of the given hash and identify algorithms
def part_e():
    hash_value = "408d890298c4963dfaac2e8b508552b9e1ee8048"
    length_bits = len(hash_value) * 4  # Each hex character represents 4 bits
    print(f"e) The hash {hash_value} is {length_bits} bits long.")
    print("   Known hashing algorithms with this length: SHA-1, RIPEMD-160")

# Run all parts
def main():
    part_a()
    part_b()
    part_c()
    part_d()
    part_e()

if __name__ == "__main__":
    main()
```
